```python
df_filtered = df[df["Smoking status"].isin(["yes", "no"])]

df_filtered["Smoking status"].unique()
```

```python

df["Smoking status"]
df_filtered["Smoking status"] = df["Smoking status"].map({"no": 0, "yes": 1})

```

```python
# 3. Scikit-learn: Model Training
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Example columns: adjust for your CSV
X = df_filtered[["Age of the patient", "Blood pressure (mm/Hg)", "Specific gravity of urine", "Albumin in urine", "Sugar in urine", "Random blood glucose level (mg/dl)", "Body Mass Index (BMI)", "Duration of diabetes mellitus (years)", "Duration of hypertension (years)", "Duration of hypertension (years)", "Cystatin C level"]]
y = df_filtered["Smoking status"]


```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print(predictions[:5])
```

```python
from sklearn.metrics import accuracy_score

# Convert probabilities to class labels using 0.5 as the threshold
predicted_labels = []
for prob in predictions:
  if prob >= 0.5:
    predicted_labels.append(1)
  else:
    predicted_labels.append(0)

print(predicted_labels)

# Now calculate accuracy
accuracy = accuracy_score(y_test, predicted_labels)
print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt

# Let's compare predictions to actual values
plt.scatter(y_test, predictions)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()
```
