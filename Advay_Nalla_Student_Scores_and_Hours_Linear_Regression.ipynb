```python
import pandas as pd
```

```python
student_scores = pd.read_csv('/content/score_updated.csv')
```

```python
student_scores
```

```python
student_scores.columns
```

```python
x = student_scores[['Hours']]
x
```

```python
y = student_scores[['Scores']]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
import matplotlib.pyplot as plt

# Make predictions
y_pred = model.predict(x)

# Scatter plot of actual data
plt.scatter(x, y, color='blue', label="Score data")

# Line plot of predictions
plt.plot(x, y_pred, color='red', linewidth=2, label="Linear Regression line")

# Labels and title
plt.xlabel("x (Hours)")
plt.ylabel("y (Scores)")
plt.title("Student Scores and Hours Linear Regression Model")
plt.legend()
plt.show()
```
